```python
# WITHOUT USING DATAFRAME
import csv
with open('customer.csv') as a:
    lines=csv.reader(a)
    for row in lines:
        print(row)
```

```python
# WITH USING DATAFRAMES
import pandas as pd
df=pd.read_csv('customer.csv')
df
```

```python
# TYPES
df.dtypes
```

```python
# FREQUENCY OF AN VALUE
df['income'].value_counts()
```

```python
# PLOTTING THE GRAPH
import matplotlib.pyplot as plt
income=df['income'].value_counts()
income.plot.bar()
```

```python
# DELETING SOME ELEMENTS FROM ROW
print(df[df.duplicated(subset=['age','income','student','creditrating','class'])])
print(df.drop_duplicates)
```

```python
#DELETING COLUMNS WITH SINGLE VALUE
for key,value in df.iteritems():
    if len(df[key].unique())==1:
        print(df[key])
        del df[key]
print(df)
```

```python
# TRANSFORMING CATEGORICAL VALUES
transform=pd.get_dummies(df['income'])
df_new=pd.concat([df,transform],axis=1)

df_new['age']=[0 if x=='senior' else 1 for x in df['age']]
df_new
```

```python
# FIRST 3 ROWS
df_new.head(3)
```

```python
# LAST 3 ROWS
df_new.tail(3)
```

```python
# FILTERING
df_new.loc[0,:]
```

```python
# ADDING NEW INSTANCES
newrow={'age':'senior','income':'high','creditrating':'fair','class':'yes','sem':2}
df_new.loc[len(df_new)]=newrow
df_new
```

```python
# TRANSFORMING VALUES TO BINS
import numpy as np
df_n=pd.DataFrame({'age_int':np.random.randint(0,100,20)})
labels=["{0}-{1}".format(i,i+9) for i in range(0,100,10)]
df_n['age_group']=pd.cut(df_n.age_int,range(0,105,10),right=False,labels=labels)
df_new=pd.concat([df_new,df_n],axis=1)
df_new
```
