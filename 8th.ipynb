```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('dta.csv')
f1=data['v1'].values
f2=data['v2'].values
X=np.array(list(zip(f1,f2)))
print("X:",X)
```

```python
plt.scatter(f1,f2,c='black')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(2)
label=kmeans.fit(X).predict(X)
print("Labels are:",labels)
centroid=kmeans.cluster_centers_
print("Centroid:",centroid)
```

```python
plt.scatter(f1,f2,c=label)
plt.show()
```

```python
plt.scatter(centroid[:,0],centroid[:,1],c='red',marker='*')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
gmm=GaussianMixture(2)
labels=gmm.fit(X).predict(X)
mean=gmm.means_
std=gmm.covariances_
print("Labels:",labels)
print("Mean:",mean)
print("Covariances:",std)
```

```python
plt.scatter(f1,f2,c=labels)
plt.show()
```
