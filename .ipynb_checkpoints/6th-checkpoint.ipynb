```python
def fract(x,y):
    if y==0:
        return 0
    return x/y
import csv
import random
import math
```

```python
def loadcsv(filename):
    lines=csv.reader(open(filename,'r'))
    a=list(lines)
    for row in range(len(a)):
        a[row]=[float(x) for x in a[row]]
    return a
```

```python
def splitdataset(a,splitratio):
    b=int(len(a)*splitratio)
    print(b)
    c=[]
    copy=list(a)
    while len(c)<b:
        index=random.randrange(len(copy))
        c.append(copy.pop(index))
    return [c,copy]
```

```python
def separatebyclass(a):
    d={}
    for i in range(len(a)):
        v=a[i]
        if v[-1] not in d:
            d[v[-1]]=[]
        d[v[-1]].append(v)
    return d
```

```python
def mean(number):
    return fract(sum(number),float(len(number)))
```

```python
def stddev(number):
    avg=mean(number)
    variance=fract(sum([pow(x-avg,2) for x in number]),float(len(number)))
    return math.sqrt(variance)
```

```python
def summarize(a):
    e=[(mean(attributes),stddev(attributes)) for attributes in zip(*a)]
    del e[-1]
    return e
```

```python
def summarizebyclass(a):
    d=separatebyclass(a)
    e={}
    for f,instances in d.items():
        e[f]=summarize(instances)
    return e
```

```python
def cp(x,mean,stddev):
    exponent=math.exp(-fract((math.pow(x-mean,2)),(2*math.pow(stddev,2))))
    return fract(1,(math.sqrt(2*math.pi*stddev)))*exponent
```

```python
def ccp(e,g):
    h={}
    for f,u in e.items():
        h[f]=1
        for i in range(len(u)):
            mean,stddev=u[i]
            x=g[i]
            print("@@",mean,stddev)
            h[f]*=cp(x,mean,stddev)
            print("Individual probability",h[f])
    return h
```

```python
def predict(e,g):
    h=ccp(e,g)
    print("@@@",h)
    j,k=None,-1
    for f,probability in h.items():
        if j is None or probability>k:
            k=probability
            j=f
    return j
```

```python
def getprediction(e,j):
    l=[]
    for i in range(len(j)):
        result=predict(e,j[i])
        l.append(result)
    return l
```

```python
def getaccuracy(j,l):
    correct=0
    for i in range(len(j)):
        if j[i][-1]==l[i]:
            correct+=1
    return fract(correct,float(len(j)))*100.00
```

```python
def naivebayes():
    filename='diabetes_6.csv'
    splitratio=0.8
    a=loadcsv(filename)
    train,j=splitdataset(a,splitratio)
    e=summarizebyclass(train)
    l=getprediction(e,j)
    accuracy=getaccuracy(j,l)
    print("Modal Summaries:",e)
    print("\n Predictions:",l)
    print("\n Accuracy:",accuracy)
naivebayes()
```
